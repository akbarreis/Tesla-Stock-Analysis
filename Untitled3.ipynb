```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # I use the library to prepare the drawing
USD = pd.read_excel('./Documents/Excelпроблем5.xlsx')#calling the excel graph 


```

```python
def lin_reg(input, output):
    plt.figure(figsize = (18,8))#I give the dimensions to the drawing
    Xi = input
    Yi = output
    n = len(Xi) #number of data 
    Xi_mean = Xi.mean()
    Yi_mean = Yi.mean()
    sum_Xi = (Xi).sum()
    sum_Yi = (Yi).sum()
    sum_XiYi = ((Xi*Yi).sum())*n #use formul linear regression
    sum_Xi_sum_Yi = sum_Xi*sum_Yi
    sumXiXi = ((Xi*Xi).sum())*n
    sum_Xi_sq = sum_Xi*sum_Xi
    a = ((sum_XiYi - sum_Xi_sum_Yi) / (sumXiXi - sum_Xi_sq))
    b = Yi_mean - (a * Xi_mean)
    return a, b
a, b = lin_reg(USD['Порядок'], USD['Курс'])
plt.title("linear regression graph with USD")#give  call title
plt.scatter(USD['Порядок'], USD['Курс'], alpha=0.93,color = 'r')
plt.plot(USD['Порядок'], (a*USD['Порядок'])+b, color='g', label=' Regression') #plotting a linear regression
plt.grid()
plt.ylabel("USD COURSE $")#I give the name to the axes
plt.xlabel("number of days per year")#I give the name to the axes
plt.legend()
plt.show()#show the graph in the drawing
```
